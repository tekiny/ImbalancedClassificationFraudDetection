```python
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns with missing values.")
        
        # Return the dataframe with missing information
        return mis_val_table_ren_columns
```

```python
#1) Baseline Model
#I will use CrossValidation with Gridsearchcv for all of my models. 
#I don't touch the test data which I splitted in the previous step.I only look at it's score after finding best trainig model
```

```python
#Baseline Model
#!!!! I use recall as a score in the gridsearch. In this model my training data is imbalanced 

scoring = {'accuracy' : make_scorer(accuracy_score), 
           'precision' : make_scorer(precision_score),
           'recall' : make_scorer(recall_score), 
           'f1_score' : make_scorer(f1_score)}

cv1 = StratifiedShuffleSplit(n_splits=5, test_size=0.2,train_size=0.8, random_state=0)    
gs = GridSearchCV(LogisticRegression(),
                  param_grid={'C': [0.01, 0.1, 1, 10, 100]},
                  scoring=scoring, refit='recall', cv=cv1)
gs.fit(train_X,train_y)
results = gs.cv_results_
print("Best params with LR as a baseline:\n{}\n".format(gs.best_params_))
print("Best cross-validation score LR as a baseline model: {:.5f}".format(gs.best_score_))
print("Test-set score with LR as a baseline model: {:.5f}".format(gs.score(test_X,test_y)))
                 
```

```python
#Undersampe
no_posit = len(df[df['Class'] == 1])
neg_indices = df2[df2.Class == 0].index
pos_indices = df[df.Class == 1].index

random_indices = np.random.choice(neg_indices,no_posit, replace=False)

under_sample_indices = np.concatenate([pos_indices,random_indices])

under_sample = df2.loc[under_sample_indices]
```

```python
y2 = under_sample['Class']
## To Review Class Imbalance

y2.value_counts().plot(kind='bar');
sns.despine();
sns.set_context('notebook');
```

```python
X_under = under_sample.loc[:,under_sample.columns != 'Class']
y_under = under_sample.loc[:,under_sample.columns == 'Class']
X_under=np.array(X_under)
y_under=np.array(y_under.pop('Class'))
```

```python
y_under.shape
```

    2)Two models with undersampling

```python
## To improve Recall -->
#2)Two models with undersampling
cv1 = StratifiedShuffleSplit(n_splits=5, test_size=0.2,train_size=0.8, random_state=0)    
gs01 = GridSearchCV(LogisticRegression(),
                  param_grid={'C': [0.01, 0.1, 1, 10, 100]},
                  scoring=scoring, refit='recall', cv=cv1)
gs01.fit(X_under,y_under)
results01 = gs01.cv_results_
print("Best params with LR model with undersampling :\n{}\n".format(gs01.best_params_))
print("Best cross-validation score with LR model with undersampling: {:.4f}".format(gs01.best_score_))
print("Test-set score with LR model with undersampling: {:.4f}".format(gs01.score(test_X,test_y)))
optimised_lr01 = gs01.best_estimator_
                 


gs02 = GridSearchCV(RandomForestClassifier(n_estimators=100),
                  param_grid={'max_features': [1,2,3]},                               
                  scoring=scoring, refit='accuracy', cv=cv1)
gs02.fit(X_under,y_under)
results02 = gs02.cv_results_
print("Best params with RF model with undersampling :\n{}\n".format(gs02.best_params_))
print("Best cross-validation score with RF model with undersampling: {:.5f}".format(gs02.best_score_))
print("Test-set score with RF model with undersampling: {:.5f}".format(gs02.score(test_X,test_y)))
optimised_rf02 = gs02.best_estimator_
                 
```

```python
#Weighted Models

#Two models with weighted class acording to their balance
#!!!!! I used "balanced" parameter which performs class weighted in LR and RF Machine learning alg 
#The “balanced” mode uses the values of y to automatically 
#adjust weights inversely proportional to class frequencies in the input data as n_samples / (n_classes * np.bincount(y))


cv1 = StratifiedShuffleSplit(n_splits=5, test_size=0.3,train_size=0.7, random_state=0)    
gs03 = GridSearchCV(LogisticRegression(class_weight = 'balanced'),
                  param_grid={'C': [0.01, 0.1, 1, 10, 100]},
                  scoring=scoring, refit='recall', cv=cv1,return_train_score=True)
gs03.fit(train_X,train_y)
results03 = gs03.cv_results_
print("Best params with LR model with Class weighting :\n{}\n".format(gs03.best_params_))
print("Best cross-validation score with LR model with Class weighting: {:.5f}".format(gs03.best_score_))
print("Test-set score with LR model with Class weighting: {:.5f}".format(gs03.score(test_X,test_y)))
optimised_lr03 = gs03.best_estimator_
                 


gs04 = GridSearchCV(RandomForestClassifier(n_estimators=100, class_weight='balanced'),
                  param_grid={'max_features': [1,2,3]},                               
                  scoring=scoring, refit='recall', cv=cv1,return_train_score=True)
gs04.fit(train_X,train_y)
results04 = gs04.cv_results_
print("Best params with RF model with Class weighting :\n{}\n".format(gs04.best_params_))
print("Best cross-validation score with RF model with Class weighting: {:.5f}".format(gs04.best_score_))
print("Test-set score with RF model with Class weighting: {:.5f}".format(gs04.score(test_X,test_y)))
optimised_rf04 = gs04.best_estimator_                 
```

```python
#Since I already splitted the test data in the preprocessing pahese now I can apply oversampling on the training data
#It is obvious that if we apply SMOTE on both test and training set, we will have a synthesized test set 
#which gives us a high accuracy that is not actually correct.

```
